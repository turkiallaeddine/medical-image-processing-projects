```python
import cv2
import matplotlib.pyplot as plt

# 1. Load image (     )
img = cv2.imread("xray.jpeg")

#      :
# img = cv2.imread("lung.png")

# 2. Convert BGR  RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 3. Convert to grayscale
gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

# 4. Resize (  )
gray_resized = cv2.resize(gray, (512, 512))

# 5.  
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original X-ray")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Grayscale X-ray")
plt.imshow(gray_resized, cmap="gray")
plt.axis("off")

plt.show()

cv2.imwrite("gray_xray.jpg", gray_resized)
```
